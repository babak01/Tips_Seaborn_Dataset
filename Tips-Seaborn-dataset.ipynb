```python
# import librarires
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import tips dataset from seaborn
"""
total_bill: Total bill (cost of the meal), including tax, in US dollars
tip: Tip (gratuity) in US dollars
sex: Sex of person paying for the meal (0=male, 1=female)
smoker: Smoker in party? (0=No, 1=Yes)
day: 3=Thur, 4=Fri, 5=Sat, 6=Sun
time: 0=Day, 1=Night
size: Size of the party
"""
tips = sns.load_dataset('tips')
```

```python
# Check the dataset
tips.head()
```

```python
fig=plt.figure(dpi=600)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Visualize the histogram and linear regression of data related 
#to the sex
fig=plt.figure(dpi=600)
sns.pairplot(tips, hue='sex', kind='reg',diag_kind='hist'
             ,diag_kws={'edgecolor':'white'})
```

```python
# Related to the Smoker
fig=plt.figure(dpi=600)
sns.set(style="ticks", color_codes=True)
sns.pairplot(tips, hue='smoker', palette='husl',
             kind='scatter', plot_kws=dict(edgecolor="white"),
             diag_kind='kde')
```
